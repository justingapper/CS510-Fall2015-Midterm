##### How does the plotted solution depend on your choice of time step size, and your choice of increment? Do you see an improvement in precision from using the higher-order integration methods?

```python
attr = attractor.Attractor()
attr.evolve(order=4)
attr.plotx()
attr.ploty()
attr.plotz()
attr.plotxy()
attr.plotyz()
attr.plotzx()
```

```python
attr = attractor.Attractor(points=100000)
attr.evolve(order=4)
attr.plotx()
attr.ploty()
attr.plotz()
attr.plotxy()
attr.plotyz()
attr.plotzx()
```

##### How does the solution depend upon the initial conditions [x0, y0, z0]? How small a change can you make while still reproducing roughly the same dynamical curves?

```python
attr = attractor.Attractor()
attr.evolve([1,1,1],4)
attr.plotx()
attr.ploty()
attr.plotz()
attr.plotxy()
attr.plotyz()
attr.plotzx()
```

```python
attr = attractor.Attractor()
attr.evolve([50,50,50],4)
attr.plotx()
attr.ploty()
attr.plotz()
attr.plotxy()
attr.plotyz()
attr.plotzx()
```

```python
attr = attractor.Attractor()
attr.evolve([50,0,0],4)
attr.plotx()
attr.ploty()
attr.plotz()
attr.plotxy()
attr.plotyz()
attr.plotzx()
```
