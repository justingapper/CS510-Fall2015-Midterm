##### How does the solution depend upon the initial conditions [x0, y0, z0]? How small a change can you make while still reproducing roughly the same dynamical curves?

```python
atr = attractor.Attractor()
atr.evolve([1,1,1])
atr.plotx()
atr.ploty()
atr.plotz()
```

```python
atr = attractor.Attractor()
atr.evolve([10,10,10])
atr.plotx()
atr.ploty()
atr.plotz()
```

```python
atr = attractor.Attractor()
atr.evolve([10,0,0])
atr.plotx()
atr.ploty()
atr.plotz()
```

```python
atr = attractor.Attractor()
atr.evolve([0,10,0])
atr.plotx()
atr.ploty()
atr.plotz()
```

##### The numbers seem to converge to the same range independent of starting values.
